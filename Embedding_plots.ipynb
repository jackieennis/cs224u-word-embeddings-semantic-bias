```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
eng_embed['Relative Female Perc'] = eng_embed['Census Female Perc'].apply(lambda x: x-(100-x))
```

```python
e = eng_embed.plot(y='Cosine Distance', x='Relative Female Perc', 
       style='o', grid="true")
e.set_ylabel('Bias')
e.set_xlabel('Percent difference of women')
e.set_title("Embedding bias in English ")
plt.show()
```
